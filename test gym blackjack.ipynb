```python
from custom_env import BlackjackEnv
```

```python
env = BlackjackEnv()

total_rewards = 0
NUM_EPISODES = 1000
```

```python
for ep in range(NUM_EPISODES):
    
    print("\nEp:",ep)
    new_state = env.reset()
    print("==============")
    env.render()
    print(new_state)
    episode_reward = 0
    while env.done == False:
        
        print("==============")
        action = env.action_space.sample()
        new_state, reward, done, desc = env.step(action)
        env.render()
        print(new_state)
        
        episode_reward += reward
        
    total_rewards += episode_reward
        
avg_reward = total_rewards / NUM_EPISODES
print(avg_reward)
```
